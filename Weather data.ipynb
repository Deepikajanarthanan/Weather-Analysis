19. Detect any unusual or extreme outliers in the dataset

```python
numeric_data = data.select_dtypes(include=['number'])
Q1 = numeric_data.quantile(0.25)
Q3 = numeric_data.quantile(0.75)
IQR = Q3 - Q1
outliers = numeric_data[(numeric_data < (Q1 - 1.5 * IQR)) | (numeric_data > (Q3 + 1.5 * IQR))]
outliers.dropna(how='all')
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

st.title("Weather Data Dashboard")

# Load Dataset
# Replace with your file path or API data
data = pd.read_csv("weather.csv")

# Convert Date/Time column
data["Date/Time"] = pd.to_datetime(data["Date/Time"])

# Extract useful time features
data["Year"] = data["Date/Time"].dt.year
data["Month"] = data["Date/Time"].dt.month
data["Day"] = data["Date/Time"].dt.day

# Sidebar Filters
st.sidebar.header("Filter Options")

selected_year = st.sidebar.multiselect(
    "Select Year",
    options=data["Year"].unique(),
    default=data["Year"].unique()
)

selected_weather = st.sidebar.multiselect(
    "Select Weather Condition",
    options=data["Weather"].unique(),
    default=data["Weather"].unique()
)

filtered_data = data[
    (data["Year"].isin(selected_year)) &
    (data["Weather"].isin(selected_weather))
]

# Show Data
st.subheader("Filtered Dataset")
st.dataframe(filtered_data)

# Temperature Trend
st.subheader("Temperature Over Time")
fig1 = plt.figure()
plt.plot(filtered_data["Date/Time"], filtered_data["Temp_C"])
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)
st.pyplot(fig1)

# Humidity vs Temperature
st.subheader("Humidity vs Temperature")
fig2 = plt.figure()
plt.scatter(filtered_data["Rel Hum_%"], filtered_data["Temp_C"])
plt.xlabel("Relative Humidity (%)")
plt.ylabel("Temperature (°C)")
st.pyplot(fig2)

# Wind Speed vs Visibility
st.subheader("Wind Speed vs Visibility")
fig3 = plt.figure()
plt.scatter(filtered_data["Wind Speed_km/h"], filtered_data["Visibility_km"])
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Visibility (km)")
st.pyplot(fig3)

# Pressure Trend
st.subheader("Pressure Over Time")
fig4 = plt.figure()
plt.plot(filtered_data["Date/Time"], filtered_data["Press_kPa"])
plt.xlabel("Date")
plt.ylabel("Pressure (kPa)")
plt.xticks(rotation=45)
st.pyplot(fig4)

# Average Weather Metrics
st.subheader("Average Weather Metrics")
avg_values = filtered_data[[
    "Temp_C", "Dew Point Temp_C", "Rel Hum_%",
    "Wind Speed_km/h", "Visibility_km", "Press_kPa"
]].mean()

fig5 = plt.figure()
avg_values.plot(kind="bar")
plt.ylabel("Average Values")
st.pyplot(fig5)

# Top Extreme Weather Records
st.subheader("Top 5 Highest Temperature Records")
top_temp = filtered_data.sort_values(by="Temp_C", ascending=False).head(5)
st.dataframe(top_temp)

st.subheader("Top 5 Lowest Temperature Records")
low_temp = filtered_data.sort_values(by="Temp_C", ascending=True).head(5)
st.dataframe(low_temp)
```
